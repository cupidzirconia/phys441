Write a Python function that rolls a die with num_sides sides a total of
num_rolls times. The function should plot a histogram of all of the rolls and
should return the average roll value.

```python
import random
import matplotlib.pyplot as plt
```

```python
def roll(num_sides, num_rolls):
    rolls = [random.randint(1, num_sides) for _ in range(num_rolls)]
    
    avg_roll = sum(rolls) / num_rolls
    
    plt.hist(rolls, bins=range(1, num_sides + 2), align="left", rwidth=0.8, density=False)
    plt.xlabel("roll value")
    plt.ylabel("frequency")
    plt.title("roll frequency histogram")
    plt.xticks(range(1, num_sides + 1))
    plt.show()
    
    return avg_roll
```

```python
num_sides = 20
num_rolls = 1000000
average = roll(num_sides, num_rolls)
print(f'Average roll value: {average:.2f}')
```

The average roll comes out to be approximately 10.5. This is expected based on (20+1)/2 = 10.5
The shape of the histogram is pretty flat. This makes sense as the probability of rolling any one number is 1/20.

```python
def max_roll(num_sides, num_rolls):
    rolls = [random.randint(1, num_sides) for _ in range(num_rolls)]
    
    max_roll = max(rolls)
    
    return max_roll
```

```python
def roll_sim(num_sides, num_rolls, num_sims):
    max_rolls = []
    
    for _ in range(num_sims):
        max_val = max_roll(num_sides, num_rolls)
        max_rolls.append(max_val)
        
    plt.hist(max_rolls, bins=range(1, num_sides + 2), align='left', rwidth=0.8, density=False)
    plt.xlabel('Maximum Roll')
    plt.ylabel('Frequency')
    plt.title(f'Maximum of Two Die Rolls ({num_sims} Times)')
    plt.xticks(range(1, num_sides + 1))
    plt.show()
```

```python
num_sides = 20
num_rolls = 2
num_sims = 1000000
roll_sim(num_sides, num_rolls, num_sims)
```

The shape of this histogram makes sense because we are taking the maximum of the two rolls. 
There is the lowest probability of returning a 1, because you would need your second roll to be a 1, which is a 1 in 20 chance. 

As the max value goes up, the less likely you are to roll above it. 
For instance, if you were to roll a 19, the only way you would return a number not equal to 19 is if you were to roll a 20, which is a 1 in 20 chance.
